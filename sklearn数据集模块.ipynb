## 1 sklearn数据集概述

### 1.1 和数据相关的概念

- 特征： 描述队形的一个属性，比如鸢尾花的花瓣长度


- 样本： 描述的一个对象，比如一朵山鸢尾花

- 特征矩阵X： [nsamples, nfeatures]构成的矩阵

- 标注向量y： 样本类别，维度为nsamples


### 1.2 sklearn数据集

在sklearn中有三种类型的数据：

- 自带小数据集：
 
 > 使用load_XXX获取；
 
 >数据在sklearn的datasets模块中；
 
 >典型的数据有：鸢尾花，波士顿房价，乳腺癌；
 
 - 自带大数据
 
  > 使用fetch_xxx从网络中下载
  
 - 通过函数构建的数据集
 
 > 使用dataset.make_XXX 构建
 > 典型的代表有：
 > > - make_blobs():产生聚类数据
 > > - make_classfication():产生分类数据
 > > - make_circle():产生分类
 > > - make_moon():产生分类数据
 > > - make_multilabel_classification():产生聚类数据
 > > - make_regression():产生回归数据
 
 
### 1.3 sklearn中数据类型

使用load_XXX 和fetch_XXX 获取的数据类型为datasets.base.Bunch类型，本质上就是一个字典类型。 包含以下属性：


- data: 特征数组，ndarray类型
- target: 标记向量，ndarray类型
- DESCR：数据集的描述
- features_names: 特征的名字
- target_name:标签名字
- image:图像格式的二进制样本数据

## 1.4 两个辅助函数

- datasets.get_data_home():获取数据下载目录名；
- datasets.clear_data_home(data_home=None):删除数据下载目录下的数据

```python
# 例2 获取自带数据

from sklearn.datasets import *

data=load_digits() # 获取数字图片信息

#产看数据类型
print(type(data))

print("属性名：")
#bunch本质是字典类型
print(data.keys())

print("——————————————————————")
print("属性值：")
#print(data.values())

print("shape:")
print(data.data.shape)
print(data.target.shape)

#可以使用matplotlib对image进行可视化

import matplotlib.pyplot as plt

# 设置图形大小(宽、高)以英寸为单位
fig = plt.figure(figsize=(6, 6))

# 设置子图形布局，如间隔之类... 
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# 对于64幅图像中的每一幅
for i in range(64):
    # 初始化子图:在8×8的网格中，在第i+1个位置添加一个子图
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    # 在第i个位置显示图像
    ax.imshow(data.images[i], cmap=plt.cm.binary, interpolation='nearest')
    # 用目标值标记图像
    ax.text(0, 7, str(data.target[i]))

# 显示图形
plt.show()
```
